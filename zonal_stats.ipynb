## About the data

**Dataset 1**

Shipping intensity in the Pacific Arctic region during August 2017. The value at each raster cell represents the toal length, in meters, of all vessel tracks within each cell. 

Resolution: 1km per pixel

**Dataset 2**

Vector data shwoing statistical areas dividing waters of the State of Alaska and the adjacent Exclusive Economic Zone (EEZ) into small units for the purpose of reporting and analyzing fishery harvest.

Format: GeoPackage (extension .gpkg). Open, non-proprietary platform-independent, self-describing format. Supports both raster and vector data. 

**Dataset 3**

Polygon showing Alaska's boundary extracted from the 2022 U.S. census TIGER shapefiles, updated CRS.

## Dissolve and explode polygons

The Kodiak data is too granular for our purposes: we want to get statistics on distance travelled on each fishing district

````python
# check number of districts and rows
print(f"number of geometries: {len(kodiak)}")
print(f"number of districts: {len(kodiak.district_name.unique())}")
````

We can aggregate the geometries based on the values of a column using the `dissolve()` method. Here we will dissolve by district name. *kinda like a geopspatial groupby**

````python
# dissolve by district
# as_index = False indicates we want to keep the 'groupby' column as a column, not index
districts = kodiak[['geometry','district_name']].dissolve(by = 'district_name', as_index = False)
districts
````

Let's take a look at the districts:

We can control the legend location adding `loc` and `bbox_to_anchor` to `legend_kwds`

It can be useful to use both of these together.

- `loc`: indicates the corner of the legend box we are aligning.
- `bbox_to_anchor`: is a tuple with coordinates of where to place the corner specifies in `loc`. Values between 0 and 1 are inside the plot (inside axes)

````python
districts.plot(column = 'district_name',
              legend = True,
              legend_kwds = {'loc':'upper left',
                            'bbox_to_anchor': (1,1)})
````

Notice some of our districts are multipolygons. To simplify our analysis (and for the purpose of this lecture) we will separte each multipolygon into individual polygons,

use`explode()` from `gpd.GeoDataFrame`:

````python
districts = districts.explode(ignore_index = True)
districts
````

## Distance in a single district

We want to calculate the total distance travelled within one of these districts.

To transfrom a geometry into a mask we can use the `rasterize()` functionfrom `rasterio.features`:


````python
# create a raster mask based on alitak polygon
alitak_mask = rasterize(shapes = [districts.loc[1,'geometry']], # a list of geometries
                       out_shape = kodiak_dist.shape, # shape of output np array (shape of raster)
                       transform = kodiak_dist.rio.transform(), # transformation of raster
                       all_touched = True, # all pixels touched by geometries will be burned in
                       ) 
print(type(alitak_mask))
alitak_mask
````

````python
print(alitak_mask.shape)
print(alitak_mask.shape == kodiak_dist.shape)
print(np.unique(alitak_mask))
plt.imshow(alitak_mask)
````

````python
districts.loc[1,'geometry']
````

````python
# apply mask to kodiak distance raster

dist_alitak = kodiak_dist * alitak_mask
dist_alitak.plot(robust = True)
````

````python
print(f"total distance in Alitak: {dist_alitak.sum().item():,}")
````

````python
# in a single cell this would be: 
alitak_mask = rasterize(shapes = [districts.loc[1,'geometry']], # a list of geometries
                       out_shape = kodiak_dist.shape, # shape of output np array (shape of raster)
                       transform = kodiak_dist.rio.transform(), # transformation of raster
                       all_touched = True, # all pixels touched by geometries will be burned in
                       ) 
dist_in_alitak = (kodiak_dist * alitak_mask).sum().item()
dist_in_alitak
````

## Distance in each district

````python
for i in districts.index:
    print(i)
````

````python
# calculate distance per district polygon
distances = []

for i in districts.index:
    mask = rasterize(shapes = [districts.loc[i,'geometry']], # a list of geometries
                       out_shape = kodiak_dist.shape, # shape of output np array (shape of raster)
                       transform = kodiak_dist.rio.transform(), # transformation of raster
                       all_touched = True, # all pixels touched by geometries will be burned in
                       ) 
    dist_in_district = (kodiak_dist * mask).sum().item()
    distances.append(dist_in_district)
    
distances
````

````python
# Add column with distance per district
districts['total_distance'] = distances
districts
````

````python
districts.plot(column = "total_distance",
              legend = True)
````
